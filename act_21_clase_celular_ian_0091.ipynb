```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} se ha encendido.")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")

# creacion de objetos (instancias)
cel1 = Celular("Samsung", "Galaxy S23", 800)
cel2 = Celular("Apple", "iPhone 15", 1000)
# llamadas a las funciones para cada objeto celular 1 y celular 2
print("celular 1:")
cel1.mostrar_info()
cel2.encender()

print("\ncelular 2:")
cel2.mostrar_info()
cel2.encender()

print("programa terminada")
print("creador de software por ian gutierrez NL= 22")
```
